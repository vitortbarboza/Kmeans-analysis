```python
##KNN Classifier

scaler=StandardScaler()
scaler.fit(df.drop('cluster_class', axis=1))
df_normalizado=scaler.transform(df.drop('cluster_class', axis=1))
df_param=pd.DataFrame(df_normalizado, columns=df.columns[:-1])
df_param.head()
```

```python
x_train, x_test, y_train, y_test = train_test_split(
    df_param, df['cluster_class'], test_size=0.3, random_state=101)
```

```python
knn=KNeighborsClassifier(n_neighbors=1)
knn.fit(x_train,y_train)
```

```python
knn_pred=knn.predict(x_test)
print(classification_report(y_test, knn_pred))

print(confusion_matrix(y_test, knn_pred))
```

```python
error_rate= []

for i in range(1,40):
    knn= KNeighborsClassifier(n_neighbors=i)
    knn.fit(x_train, y_train)
    pred=knn.predict(x_test)
    error_rate.append(np.mean(pred!=y_test))
```

```python
sns.set_style('darkgrid')
plt.figure(figsize=[14,8])
plt.plot(range(1,40), error_rate, color='blue',
        linestyle='dashed', marker='o')
plt.xlabel('K')
plt.ylabel('Error Rate')
```
